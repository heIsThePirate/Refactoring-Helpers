```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from math import pi

N = 256

x = np.arange(0, 2*pi, 2*pi/N)
y = np.arange(0, 2*pi, 2*pi/N)

l = []

with open('./rho/rho0.txt') as f:
    for line in f:
        l = line.split()

data = np.array([float(num) for num in l])
data = data.reshape((N, N))

plt.pcolor(x, y, data, cmap='jet');
```
